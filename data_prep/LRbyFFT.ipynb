```python
import numpy as np
import os
import glob
from PIL import Image
import random

outdir = '/content/gdrive/My Drive/Colab Notebooks/MRI BRAIN image SR/mDCSRN_WGANGP_v2/HCP_1200_NPY_LR_' #HCP_NPY를 위한 폴더 경로 생성 
if not os.path.exists(outdir):
    os.makedirs(outdir) # 생성 
    

files = glob.glob("/content/gdrive/My Drive/Colab Notebooks/MRI BRAIN image SR/mDCSRN_WGANGP_v2/HCP_1200_NPY/*_3T_T1w_MPR1.npy") 
for filepath in files:
    img_npy = np.load(filepath)
    img = np.array(img_npy,dtype=np.float32)

    alpha = random.random()

    imgfft = (1-alpha)*np.fft.fftn(img) + alpha*np.fft.fftshift(img)
    
  
    degree = 160
    x_center = imgfft.shape[0] // 2 
    y_center = imgfft.shape[1] // 2
    z_center = imgfft.shape[2] // 2
    imgfft[x_center-degree : x_center+degree, y_center-degree : y_center+degree, z_center-degree : z_center+degree] = 0
  
    imgifft = np.fft.ifftn(imgfft)
  
    img_out = abs(imgifft)
    #imgifft = np.real(imgifft)
    #imgifft -= np.amin(imgifft)
    #imgifft /= np.amax(imgifft)
    #imgifft *= 255 
    #img_out = imgifft.round().astype(np.uint8) #img_out is the result of getting LR 
    #img_out = gaussian_filter(img_out, sigma=2)

    print('  Data shape is ' + str(img_out.shape) + ' .')
    filename_ = filepath.split('/')[-1].split('.')[0]
    filename = filename_ +'_LR'+ '.npy'
    file1 = os.path.join(outdir, filename) #파일명을 수정하여 outdir path 설정
    np.save(file1, img_out) #file에 있는 데이터를 file1에 save  
    print('File ' + filename_ + ' is saved in ' + file1 + ' .') # type 변경하여 저장 성공 여부 출력
```
