```python
import numpy as np
import os
import glob
import nibabel

outdir = '/content/gdrive/My Drive/Colab Notebooks/MRI BRAIN image SR/mDCSRN_WGANGP_v2/HCP_1200_NPY' #HCP_NPY를 위한 폴더 경로 생성 
if not os.path.exists(outdir): 
    os.makedirs(outdir) # 생성 
    
files = glob.glob('/content/gdrive/My Drive/Colab Notebooks/MRI BRAIN image SR/mDCSRN_WGANGP_v2/HCP_1200/59*/*/*/*/*.nii.gz')
for filepath in files:
    print(filepath)
        #file = nibabel.load(filepath).get_data().transpose(1, 0, 2) 
        #print(filepath)
    file_ = nibabel.load(filepath)
    file = np.array(file_.dataobj)
    print('  Data shape is ' + str(file.shape) + ' .')
    filename_ = filepath.split('/')[-1].split('.')[0]
    filename = filename_ + '.npy'
    file1 = os.path.join(outdir, filename) #파일명을 수정하여 outdir path 설정
    np.save(file1, file) #file에 있는 데이터를 file1에 save  
    print('File ' + filename_ + ' is saved in ' + file1 + ' .') # type 변경하여 저장 성공 여부 출력
```
